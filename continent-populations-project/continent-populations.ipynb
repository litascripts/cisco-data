#### Line Plot

- Create a line plot with year on the x-axis and population on y-axis for each `region`.

```python
# quick data cleaning
df = indicators[['year', 'region', 'population']]   # select columns of interest
df = df.query('region != "world"')  # only include 4 regions
df
```

```python
# Pivot the DataFrame for the line plot
df_pivot = df.pivot(index='year', columns='region', values='population').reset_index()
df_pivot.columns.name = None  # remove index name
df_pivot
```

```python
def add_end_labels(df, x, column_names, alpha=1, fontweight='normal'):
    for column_name in column_names:
        y = df[column_name].iloc[-1]
        label = f'  {column_name}'
        plt.text(x, y, label, va="center", alpha=alpha, fontweight=fontweight)
```

```python
# YOUR CODE HERE (add more cells as needed)

# Convert population to billions and round
df["pop_billions"] = (df["population"] / 1e9).round()

# Filter years 2000 and 2100
df_2000 = df[df["year"] == 2000]
df_2100 = df[df["year"] == 2100]
```

```python
# horizontal bar plot

# Set up side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)

# Plot for year 2000
axes[0].barh(df_2000["region"], df_2000["pop_billions"], color='skyblue')
axes[0].set_title("Population by Region in 2000")
axes[0].set_xlabel("Population (Billions)")
axes[0].invert_yaxis()

# Plot for year 2100
axes[1].barh(df_2100["region"], df_2100["pop_billions"], color='orange')
axes[1].set_title("Projected Population by Region in 2100")
axes[1].set_xlabel("Population (Billions)")
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Line Plot of Population Over Time by Region
# Pivot table for line plot
df_pivot = df.pivot(index='year', columns='region', values='population') / 1e9  # convert to billions

# Plot
plt.figure(figsize=(10, 6))
for region in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[region], label=region)

plt.title("Population Over Time by Region")
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.legend(title="Region")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter for 2100 and sort
df_2100 = df[df["year"] == 2100].copy()
df_2100["pop_billions"] = (df_2100["population"] / 1e9).round(1)

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.barh(df_2100["region"], df_2100["pop_billions"], color="steelblue")
plt.xlabel("Population (Billions)")
plt.title("Where Will Everyone Live in 2100?")
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
