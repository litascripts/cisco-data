# The Ocean's Longest Diving Animals

How long can air-breathing animals hold their breath while diving? Which animals can stay submerged for the longest time?

The file `longest-diving-animals.csv` contains the recorded durations, in minutes, of the longest dives ever recorded in the scientific literature for different species of marine mammals.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('longest-diving-animals.csv')
df.head()
```

### Project Ideas

- Create horizontal bar plots to show the longest diving animals, either for each category or within a category.

- Follow the Pro Tips.

- Convert dive durations from minutes to hours for easier understanding - 2.5 hours is clearer than 150 minutes.

- Add a reference bar to help readers build intuition about the durations. 

Side Note: The longest time a person has held their breath underwater is 24 minutes 37 seconds. This was done at the surface of a pool, not while freediving, so it might not directly comparable.


```python
# YOUR CODE HERE (add additional cells as needed)

# Convert minutes to hours
df["duration_hours"] = df["duration"] / 60

import matplotlib.pyplot as plt

# Sort by duration
df_sorted = df.sort_values("duration_hours", ascending=True)

# Plot
plt.figure(figsize=(8, 5))
plt.barh(df_sorted["animal"], df_sorted["duration_hours"], color="navy")

# Reference bar (world record human breath-hold = 24 min 37 sec  0.41 hours)
plt.axvline(0.41, color='red', linestyle='--', label="Human record: 0.41 hrs")

# Labels
plt.xlabel("Dive Duration (hours)")
plt.title("Longest Diving Penguins (in hours)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Filter for penguins only
penguins = df[df["category"] == "penguins"].copy()

# Convert duration to hours
penguins["duration_hours"] = penguins["duration"] / 60

# Sort values for better bar layout
penguins_sorted = penguins.sort_values("duration_hours", ascending=True)

# Plot
plt.figure(figsize=(8, 5))
bars = plt.barh(penguins_sorted["animal"], penguins_sorted["duration_hours"], color="#86bf91")

# Add reference line for human breath-hold (24m37s  0.41 hours)
plt.axvline(0.41, color="red", linestyle="--", linewidth=2, label="Human Record (24m37s)")

# Labels and layout
plt.xlabel("Dive Duration (hours)")
plt.title("Longest Diving Penguins")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

The Emperor Penguin stands out with over 0.5 hours (32 minutes) dive time.

Other penguins dive significantly less  Rockhopper around 3 minutes.

The human world record breath-hold (24 mins) is surprisingly close to some of these animals  but not while diving!
