# Silent Spring

When the pesticide DDT was made available to the public in 1944, it was seen as a miracle chemical in the battle against insect pests. How did the worldwide availability of DDT coincide with word trends for insect words?

```python
# FOR GOOGLE COLAB ONLY.
# Uncomment and run the code below. A dialog will appear to upload files.
# Upload 'animal-word-trends-silent-spring.csv' and 'sparrowhawk-eggshell-data.csv'.

# from google.colab import files
# uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('animal-word-trends-silent-spring.csv')
df.head()
```

```python
def plot_insect_word_trend( animal ):
    trend = df.query('word == @animal')
    plt.figure(figsize=(5, 3))
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency per million')
    plt.title(f'Word usage for "{animal}" and the year DDT was made available')
    plt.axvline(1944, color='orange', ls='--')
```

Explore word trends for insects relative to 1944.

```python
plot_insect_word_trend('flea')
# try flea, mosquito, fly, bedbug, louse
```

#### Impact of DDT

Early marketing for DDT claimed it only killed insect pests and did not affect other animals. The following python script might help shed light on this claim.

```python
egg = pd.read_csv('sparrowhawk-eggshell-data.csv')
plt.figure(figsize=(5, 3))
plt.plot(egg['year'],egg['avg_thickness'], color='C5')
plt.xlabel('Year')
plt.ylabel('Thickness Index')
plt.title('Average Eggshell Thickness')
plt.axvline( 1944, color='magenta', ls='--', label='DDT made available to public' )
plt.show()
```

#### Rachel Carson

In 1962, Rachel Carson published the book Silent Spring, which showed that the toxins killing insects were also harming the rest of the food web. This book began to raise public awareness of DDT's toxicity, which eventually lead to bans on DDT.

In particular, birds of prey such as peregrine falcons, and songbirds such as robins, were highly affected by widespread use of DDT.

Use the function below to explore the word usage levels for these birds relative to the 1962 book.


```python
def plot_bird_word_trend( animal ):
    trend = df.query('word == @animal')
    plt.figure(figsize=(5, 3))
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency per million')
    plt.title(f'Word usage for "{animal}" and the year Silent Spring published')
    plt.axvline(1962, color='orange', ls='--')
```

```python
plot_bird_word_trend('robin')
# try robin, hawk, peregrine, osprey, sparrowhawk, condor
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load eggshell data
df_egg = pd.read_csv("sparrowhawk-eggshell-data.csv")

# Plot
plt.figure(figsize=(8, 4))
plt.plot(df_egg["year"], df_egg["avg_thickness"], marker='o', color='saddlebrown')
plt.title("Sparrowhawk Eggshell Thickness Over Time")
plt.xlabel("Year")
plt.ylabel("Avg Shell Thickness (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load word frequency data
df_words = pd.read_csv("animal-word-trends-silent-spring.csv")

# Separate by species category
birds = ["condor"]
insects = ["fly", "bedbug", "flea"]

# Plot
plt.figure(figsize=(10, 5))
for word in birds:
    df = df_words[df_words["word"] == word]
    plt.plot(df["year"], df["frequency"], label=word, linestyle='-', color='darkblue')

for word in insects:
    df = df_words[df_words["word"] == word]
    plt.plot(df["year"], df["frequency"], label=word, linestyle='--', alpha=0.7)

plt.title("Word Frequency Trends: Birds vs Insects")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

- Birds (e.g., condor) might show increasing attention during environmental awareness peaks (e.g., 1962 Silent Spring).
- Insects like flies/fleas often maintain higher baseline mentions due to nuisance, disease, or everyday context.
- Drop in eggshell thickness correlates with known DDT usage periods (1940s1970s).

Birds (e.g., "condor")

Lower overall frequency: Bird-related words like "condor" appear much less frequently in literature compared to insect words.

Spike around awareness periods: Mentions of birds may show distinct increases during major environmental events (e.g., the release of Silent Spring in 1962), reflecting public concern about bird populations and environmental health.

Associated with conservation: Birds are often used as indicators of ecological well-being, and their mentions may rise in the context of habitat loss, pesticide use (like DDT), and wildlife protection.

Insects (e.g., "fly", "flea", "bedbug")

Much higher baseline frequency: Words like "fly" and "flea" occur much more often in texts, likely because they are common, visible pests in everyday life.

More consistent trends: Insect-related word usage tends to be steady or cyclical, driven by general nuisance, disease outbreaks, or seasonal factors rather than conservation topics.

Often negative associations: Mentions often arise in negative or domestic contexts (e.g., infestations, discomfort, hygiene).

How did DDT impact animal words?

 "How DDT Changed the Way We Talk About Animals"
In the early 1900s, few people thought twice about the delicate balance of nature. Insects buzzed in the background of everyday lifeflies, fleas, and bedbugs were simply pests to be swatted away. Birds, like sparrowhawks and condors, soared quietly through literature, rarely making headlines.

But that would soon change.

 The Rise of DDT: A Double-Edged Sword
By the 1940s and 50s, DDT emerged as a miracle chemical. It saved lives by killing disease-carrying mosquitoes and cleared homes of insects. If you looked at the books and articles of the time, insect-related words remained front and center, reflecting both fascination and frustration. Words like fly and flea buzzed across the pagesjust as relentless as the real thing.

But there was a silent cost.

 The Birds That Vanished from the Sky
Birds began to disappear.

Eggshells became dangerously thin. Sparrowhawks stopped hatching. Populations of condors and eagles plummeted. Yet the danger wasn't widely knownuntil one voice changed everything.

In 1962, Rachel Carson published Silent Spring. The book unveiled the hidden toll of pesticides like DDT, and suddenly, birds had a voice in our cultural conversation.

If you analyze the word frequency trends, a pattern emerges: bird-related words began rising. Mentions of species like condor surged in books and newspapers. For the first time, people werent just noticing birdsthey were worried about them.

 Insects Stayed in the Spotlightfor Different Reasons
In contrast, insects never left the spotlightbut not because people missed them.

Instead, the language around insects reflected an ongoing battle. Despite DDTs early effectiveness, some pests bounced back, stronger than ever. Terms like bedbug and flea continued to fill pagesnot because people wanted to save them, but because they were still a nuisance.

 A Shift in Perspective
So, how did DDT change the way we talk about animals?

It amplified the voices of birdsboth literally and figurativelyas people realized the damage being done.

It kept insect-related words prevalent, though the narrative stayed centered on control and eradication.

And it marked a turning point in public consciousness, where conservation began to rival convenience.
