### Possible Explanatory Mechanism 1: Coral Bleaching

Coral bleaching occurs when coral reefs lose their vibrant colors and turn white, primarily from warm ocean temperatures. This stress causes corals to expel the essential algae, zooxanthellae, which provide them with food and color. Prolonged warm conditions can lead to coral starvation and death.

Researchers studied coral bleaching events at 100 locations across 54 countries from 1980 to 2016. The file `bleaching-reefs.csv` documents these events. Bleaching is classified as severe (`S`) if over 30% of coral colonies are affected across large areas and moderate (`M`) if 1-30% of colonies are affected at multiple sites at that location.

```python
bleaching_reefs = pd.read_csv('bleaching-reefs.csv')
bleaching_reefs
```

```python
# Count How Many Locations Experienced Coral Bleaching Each Year

import pandas as pd
import matplotlib.pyplot as plt

# Load the bleaching dataset
bleaching_df = pd.read_csv("bleaching-reefs.csv")

# Count number of locations per year
bleaching_events_per_year = bleaching_df.groupby('year')['location'].count()

# Bar chart of bleaching events per year
plt.figure(figsize=(12, 5))
bleaching_events_per_year.plot(kind='bar', color='coral')
plt.title('Number of Coral Bleaching Events per Year')
plt.xlabel('Year')
plt.ylabel('Number of Locations with Bleaching')
plt.tight_layout()
plt.show()
```

```python
print(bleaching_df['severity'].unique())
```

```python
# Bar Chart of Severe Bleaching Events per Year

# Strip whitespace from the 'severity' column
bleaching_df['severity_clean'] = bleaching_df['severity'].str.strip()

# Filter for severe cases
severe_bleaching_df = bleaching_df[bleaching_df['severity_clean'] == 'S']

# Group by year and count
severe_events_per_year = severe_bleaching_df.groupby('year')['location'].count()

# Plot
if severe_events_per_year.empty:
    print("No severe events found after cleaning.")
else:
    plt.figure(figsize=(12, 5))
    severe_events_per_year.plot(kind='bar', color='darkred')
    plt.title('Number of Severe Coral Bleaching Events per Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Locations with Severe Bleaching')
    plt.tight_layout()
    plt.show()
```

```python
# load data

import pandas as pd
import matplotlib.pyplot as plt
from linear_model import LinearModel

# Load CSVs
fish_id = pd.read_csv("reef-fish-identification-book.csv")
word_trends = pd.read_csv("animal-word-trends-coral-conundrum.csv")

# Confirm expected columns exist
print("fish_id columns:", fish_id.columns.tolist())      # Expect: ['word', 'num_species_in_book']
print("word_trends columns:", word_trends.columns.tolist())  # Expect: ['year', 'word', 'frequency']

# Normalize names
fish_id["word"] = fish_id["word"].str.lower().str.strip()
word_trends["word"] = word_trends["word"].str.lower().str.strip()

# Filter word usage for 1998
word_1998 = word_trends[word_trends["year"] == 1998][["word", "frequency"]]

# Merge datasets on 'word'
merged_df = pd.merge(fish_id, word_1998, on="word", how="inner")

# Preview
print(merged_df.head())
```

```python
# Model fit
x = merged_df["num_species_in_book"]
y = merged_df["frequency"]

model = LinearModel("Fish Species vs. Word Frequency")
model.fit(x, y)
model.print_model_info()

# Scatter plot with model line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label="Data Points", color="teal")
model.plot_model(x.min(), x.max(), color="black")

plt.title("Species Count vs. Word Frequency (1998)")
plt.xlabel("Number of Species in Book")
plt.ylabel("Word Frequency per Million")
plt.legend()
plt.tight_layout()
plt.show()
```

🔍 Questions to Consider: Interpretation

📊 What patterns emerge?

If the slope is positive and R² is moderate to high, it suggests:

Fish that have more species listed in the guide are mentioned more frequently in written texts — possibly due to higher visibility or importance.

📖 Does the guide influence word frequency?


A positive relationship may imply that the guide shaped public and scientific language, possibly making certain species names more common in writing.

📉 How might trends change in the internet era?
As print guides become less dominant, digital databases and apps (like iNaturalist, FishBase, or social media) likely influence what species names appear in publications or online discussions.

This could lead to:

A broader variety of names used (due to community input),

Or, possibly, a decline in overall fish-name frequency in academic texts as popular attention shifts away.

```python
# YOUR CODE HERE (add additional cells as needed)

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
word_trends = pd.read_csv("animal-word-trends-coral-conundrum.csv")

# Normalize fish names
word_trends["word"] = word_trends["word"].str.lower().str.strip()

# Plot Word Frequency Trends for Popular Fish (2000–2020)
# Filter for popular fish names in your guide
popular_fish = ['angelfish', 'butterflyfish', 'parrotfish', 'grouper', 'snapper']
trend_subset = word_trends[word_trends['word'].isin(popular_fish)]

# Filter for recent years only (2000–2020)
trend_subset = trend_subset[trend_subset["year"] >= 2000]

# Plot
plt.figure(figsize=(12, 6))
for fish in popular_fish:
    fish_data = trend_subset[trend_subset["word"] == fish]
    plt.plot(fish_data["year"], fish_data["frequency"], label=fish)

plt.axvline(2005, color='red', linestyle='--', label="2005 (Fish ID Book)")
plt.title("Word Frequency Trends of Popular Reef Fish (2000–2020)")
plt.xlabel("Year")
plt.ylabel("Frequency per Million Words")
plt.legend()
plt.tight_layout()
plt.show()


# Add SCUBA Certifications for Comparison
# Load SCUBA certification data
scuba_df = pd.read_csv("new-scuba-certifications.csv")

# Filter for 2000–2020
scuba_df = scuba_df[(scuba_df["year"] >= 2000) & (scuba_df["year"] <= 2020)]

# Plot alongside word frequency for context
plt.figure(figsize=(12, 5))
plt.plot(scuba_df["year"], scuba_df["num_new_certifications"], label="SCUBA Certifications", color='teal', linewidth=2)
plt.axvline(2005, color='red', linestyle='--', label="2005 (Fish ID Book)")

plt.title("New SCUBA Certifications (2000–2020)")
plt.xlabel("Year")
plt.ylabel("Certifications")
plt.legend()
plt.tight_layout()
plt.show()
```
