Can books change cultures?

The 1975 publication of the book Animal Liberation by Peter Singer is often credited with sparking an animal rights movement. Did it? Let's look at word usage data to explore this claim? We will focus on primates used in biomedical research.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('animal-word-trends-primate-research.csv')
df.head()
```

```python
def plot_word_trend( animal ):
    trend = df.query('word == @animal')
    plt.figure(figsize=(5, 3))
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency per million')
    plt.title(f'Word usage for "{animal}" and year Animal Liberation book published')
    plt.axvline(1975, color='orange', ls='--')
```

Explore word trends for primates used in research.

```python
plot_word_trend('primate')
# try primate, marmoset, baboon, rhesus
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs
word_trends = pd.read_csv("animal-word-trends-primate-research.csv")
research_animals = pd.read_csv("number-of-research-animals-per-year.csv")
```

```python
# plot

plt.figure(figsize=(10, 5))

# Plot each primate word
for word in word_trends['word'].unique():
    df_word = word_trends[word_trends['word'] == word]
    plt.plot(df_word['year'], df_word['frequency'], label=word)

# Highlight 1975 (book publication year)
plt.axvline(x=1975, color='gray', linestyle='--', label="1975: Animal Liberation")

plt.title("Primate Word Usage Over Time")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot Number of Primates Used in Research

plt.figure(figsize=(8, 5))
plt.plot(research_animals['year'], research_animals['primates'], marker='o', label="Primates Used")

# Highlight 1975
plt.axvline(x=1975, color='gray', linestyle='--', label="1975: Animal Liberation")

plt.title("Number of Primates Used in Research Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Primates")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
before = research_animals[research_animals['year'] < 1975]['primates'].mean()
after = research_animals[research_animals['year'] > 1975]['primates'].mean()

print(f"Average before 1975: {before:.0f}")
print(f"Average after 1975: {after:.0f}")
print(f"Percent change: {((after - before) / before) * 100:.2f}%")
```

Explore potential explanations for the word usage trends.
1. Ethical Influence from Animal Liberation (1975)

Sparked increased discourse on animal rights, leading to more mentions of primates in ethical debates and literature.

2. Rise in Biomedical Research

Scientific use of primates in neuroscience, immunology, and AIDS research boosted term frequency for species like macaques and rhesus.

3. Policy and Legal Changes

Amendments to animal welfare laws required more documentation and discussion of animal use, increasing word mentions.

4. Media and Public Attention

News coverage and activism highlighted primate research, spreading the terms into wider discourse beyond science.

5. Shift in Scientific Language

Greater moral awareness led researchers to refer to specific species rather than general terms like monkey.

6. Cultural Lag in Institutions

Ethical ideas took time to influence academic and regulatory institutions, causing a delayed rise in word frequency.

7. Adoption of Alternative Methods

As new testing models emerged, the practical use of primates declined, possibly reducing related terminology in later years.

What conclusions can you draw about the claim that the book Animal Liberation sparked an animal rights movement?

Animal Liberation did spark an animal rights movement, particularly by reshaping public values, academic discourse, and long-term policy.
However, the impact on actual research practices (e.g., reduction in animal use) was gradual and indirect, showing the difference between ideological influence and institutional change.

Did a book spark the animal rights movement?

Yes, but with important context.

Animal Liberation did not create the animal rights movement from nothing but it was the key spark that ignited it as a modern, organized, and philosophically grounded movement.
