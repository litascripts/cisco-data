```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
# YOUR CODE HERE (add more cells as needed)
# What are the 10 largest islands in the tropics?

largest_tropic_islands = df[df['climate'] == 'tropics'].nlargest(10, 'area')
print("10 largest islands in the tropics:")
print(largest_tropic_islands)
```

```python
# What are the largest islands in each region?
largest_by_region = df.loc[df.groupby('region')['area'].idxmax()]
print("\nLargest island in each region:")
print(largest_by_region)
```

```python
# Create a line graph with area on the y-axis and rank on the x-axis. The data should be ordered by rank, from largest to smallest.
import matplotlib.pyplot as plt

# Sorting by rank descending (largest to smallest rank number)
df_sorted = df.sort_values(by='rank', ascending=False)
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['rank'], df_sorted['area'], marker='o')
plt.xlabel('Rank (descending)')
plt.ylabel('Area')
plt.title('Island Area by Rank')
plt.grid(True)
plt.show()
```

```python
# What islands are composed of multiple countries?
multi_country_islands = df[df['countries'].str.contains(",")]
print("\nIslands with multiple countries:")
print(multi_country_islands)
```
